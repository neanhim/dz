# Тема множество
### Задание №2
Даны три множества a,b и с. Необходимо выполнить все изученные виды бинарных операций над всеми комбинациями множеств.

```python
from pprint import pprint

def union(a = [], b = []):
    c = set()
    for _ in a:
        c.add(_)
    for _ in b:
        if _ not in c:
            c.add(_)
    return c

def product(a = [], b = []):
    return set([_ for _ in a if _ in b])

def razn(a = [], b = []):
    return set([_ for _ in a if _ not in b])

def simm_razn(a = [], b = []):
    return set(union(razn(a,b), razn(b,a)))

def dek_prod(a = [], b = []):
    res = set()
    for first in a:
        for second in b:
            res.add((first,second))
    return res
```

```python
a = set([1, 3, 5])
b = set([2, 3, 4])
c = set([1, 5, 4, 2, 7])

pprint(f'объединение a и b: {union(a,b)}')
pprint(f'объединение a и с: {union(a,c)}')
pprint(f'объединение b и с: {union(b,c)}')
pprint(f'пересечение a и b: {product(a,b)}')
pprint(f'пересечение a и c: {product(a,c)}')
pprint(f'пересечение b и c: {product(b,c)}')
pprint(f'разность a и b: {razn(a,b)}')
pprint(f'разность a и c: {razn(a,c)}')
pprint(f'разность b и c: {razn(b,c)}')
pprint(f'разность b и a: {razn(b,a)}')
pprint(f'разность c и a: {razn(c,a)}')
pprint(f'разность c и b: {razn(c,b)}')
pprint(f'симметрическая разность a и b: {simm_razn(a,b)}')
pprint(f'симметрическая разность a и c: {simm_razn(a,c)}')
pprint(f'симметрическая разность b и c: {simm_razn(b,c)}')
pprint(f'декартово произведение a и b: {dek_prod(a,b)}')
pprint(f'декартово произведение a и c: {dek_prod(a,c)}')
pprint(f'декартово произведение b и c: {dek_prod(b,c)}')
pprint(f'декартово произведение b и a: {dek_prod(b,a)}')
pprint(f'декартово произведение c и a: {dek_prod(c,a)}')
pprint(f'декартово произведение c и b: {dek_prod(c,b)}')
```

# Тема 3. Последовательность

### Задание №3

```python
import numpy as np

prev = 0
znam = np.array([])
n = 1
res = 1e7
eps = 1e-7

while True:
    znam = np.append(znam, np.log(n))
    res = np.log(n) - znam.mean()
    res = np.exp(res)
    if abs(res - prev) <= eps:
        break
    prev = res
    n += 1

print(f'Предел равен {np.round(res,7)}')
```
