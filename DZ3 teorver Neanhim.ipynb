```python
import numpy as np
from math import factorial

def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

### Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
pay = np.array([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])
mean_pay = pay.sum()/len(pay)
print(f"Среднее арифметическое = {mean_pay}")
sq_pay = sum((pay - mean_pay) ** 2)
print(f"Смещенная дисперсия = {sq_pay/len(pay)}, несмещенная дисперсия = {np.around(sq_pay/(len(pay)-1), decimals = 3)}")
print(f"СКО = {np.around(np.sqrt(sq_pay/len(pay)), decimals = 3)}")
```

### В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом 2 мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

Всего способов достать мячи из ящиков
$$C^2_{8} \cdot C^4_{12}$$
Способов не достать из первого ящика белых мячей:
$$C^2_{3} \cdot C^3_{5} \cdot C^1_{7}$$
Способов  достать из первого ящика 1 белый мяч, а остальные из 2-го:
$$C^1_{3} \cdot C^1_{5} \cdot C^2_{5} \cdot C^2_{7}$$
Способов  достать из первого ящика 2 белых мяча, а остальные из 2-го:
$$C^2_{5} \cdot C^1_{5} \cdot C^3_{7}$$
Суммарная вероятность:
$$\frac{C^2_{3} \cdot C^3_{5} \cdot C^1_{7} + C^1_{3} \cdot C^1_{5} \cdot C^2_{5} \cdot C^2_{7} + C^2_{5} \cdot C^1_{5} \cdot C^3_{7}}{C^2_{8} \cdot C^4_{12}}$$


```python
p = (combinations(3, 2)*combinations(7, 1)*combinations(5, 3)+combinations(3, 1)*combinations(5, 1)*combinations(5, 2)*combinations(7, 2)+combinations(5, 2)*combinations(5, 1)*combinations(7, 3))/(combinations(8, 2)*combinations(12, 4))
print(f"Вероятность, что три мяча - белые:  p = {p : .3f}")
```

### На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.

```python
p1 = 0.9
p2 = 0.8
p3 = 0.6
print(f"Вероятность того, что стрелял первый спортсмен = {p1/(p1 + p2 + p3):.3f}")
print(f"Вероятность того, что стрелял второй спортсмен = {p2/(p1 + p2 + p3):.3f}")
print(f"Вероятность того, что стрелял третий спортсмен = {p3/(p1 + p2 + p3):.3f}")
```

### В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?

На факультеты А и B поступило по N человек, на С - 2N человек. Студентов, сдавших сессию: на А - 0.8N, на В - 0.7N, на С - 0.9*2N. Всего - 2.4N. Вероятность того, что сдавший ~~бездельник~~ студент учится на факультете Х - число сдавших студентов факультета Х  делённое на общее число сдавших студентов.

```python
p1 = 0.8
p2 = 0.7
p3 = 0.9 * 2
print(f"Вероятность того, что сдавший студент учится на факультете А = {p1/(p1 + p2 + p3):.3f}")
print(f"Вероятность того, что сдавший студент учится на факультете В = {p2/(p1 + p2 + p3):.3f}")
print(f"Вероятность того, что сдавший студент учится на факультете С = {p3/(p1 + p2 + p3):.3f}")
```

### Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25. Какова вероятность того, что в первый месяц выйдут из строя: а). все детали б). только две детали в). хотя бы одна деталь г). от одной до двух деталей?

Выход из строя деталей - независимые события.

```python
p1 = 0.1
p2 = 0.2
p3 = 0.25
print(f"Вероятность того, что выйдут из строя три детали = {p1 * p2 * p3:.3f}")
```

Если должны выйти из строя две детали, то это либо 1 и 2, 2 и 3 или 1 и 3. Просуммируя вероятности выхода из строя этих деталей найдем искомую вероятность. 

```python
print(f"Вероятность того, что выйдут из строя 2 детали = {p1 * p2 * (1 - p3) + p1 * p3 * (1 - p2) + p3 * p2 * (1 - p1):.3f}")
```

```python
print(f"Вероятность того, что выйдет из строя хотя бы 1 деталь = {1 - (1 - p1)*(1 - p2)*(1 - p3):.3f}")
```

От одной до двух деталей выйдет из строя во всех случаях, кроме выхода из строя всех деталей или не выхода ни одной.

```python
print(f"Вероятность того, что выйдет из строя от одной до двух деталей = {1 - (1 - p1)*(1 - p2)*(1 - p3) - p1 * p2 * p3:.3f}")
```
